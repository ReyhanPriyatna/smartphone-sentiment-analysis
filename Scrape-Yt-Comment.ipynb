### **Sentiment Analysis**

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline
import os
```

```python
data = pd.read_csv('comment_data.csv')
data.columns
data
```

```python
data1=data.drop(['Unnamed: 0', 'Likes', 'Time', 'user'],axis=1)
data1
```

```python
# Import functions for data preprocessing & data preparation
from sklearn.preprocessing import LabelEncoder
from sklearn.utils import resample
from sklearn.feature_extraction.text import CountVectorizer
from nltk.sentiment.vader import SentimentIntensityAnalyzer

from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from nltk.stem import PorterStemmer, LancasterStemmer
from nltk.stem.snowball import SnowballStemmer
from nltk.corpus import stopwords
from nltk.corpus import wordnet
import string
from string import punctuation
import nltk
import re
```

```python
nltk.download('vader_lexicon')
sentiments = SentimentIntensityAnalyzer()
data1["Positive"] = [sentiments.polarity_scores(i)["pos"] for i in data1["Comment"]]
data1["Negative"] = [sentiments.polarity_scores(i)["neg"] for i in data1["Comment"]]
data1["Neutral"] = [sentiments.polarity_scores(i)["neu"] for i in data1["Comment"]]
data1['Compound'] = [sentiments.polarity_scores(i)["compound"] for i in data1["Comment"]]
score = data1["Compound"].values
sentiment = []
for i in score:
    if i >= 0.05 :
        sentiment.append('Positive')
    elif i <= -0.05 :
        sentiment.append('Negative')
    else:
        sentiment.append('Neutral')
data1["Sentiment"] = sentiment
data1.head()
```

```python
data2=data1.drop(['Positive','Negative','Neutral','Compound'],axis=1)
data2.head()
```

```python
def clean_text(text):
    '''Make text lowercase, remove text in square brackets,remove links,remove punctuation
    and remove words containing numbers.'''
    text = str(text).lower()
    text = re.sub('rt', '', text)
    text = re.sub('\[.*?\]', '', text)
    text = re.sub('https?://\S+|www\.\S+', '', text)
    text = re.sub('<.*?>+', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\n', '', text)
    text = re.sub('\w*\d\w*', '', text)
    text = re.sub(' +', ' ', text)
    return text
```

```python
data_copy = data2.copy()
data_copy['Comment'] = data_copy['Comment'].apply(lambda x: clean_text(x))
```

```python
from nltk.tokenize import word_tokenize
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory

def tokenizingText(text):
    text = word_tokenize(text)
    return text

def filteringText(text):
    stop_factory = StopWordRemoverFactory()
    stopword_remover = stop_factory.create_stop_word_remover()
    
    listStopwords = stop_factory.get_stop_words()
    listStopwords.extend(['bang', 'david','gw','gua','dengan', 'ia', 'bahwa', 'oleh', 'yg', 'aku', 'kamu', 'yang', 'dan', 'di', 'dg', 'rt', 'dgn', 'ny', 'd', 'klo', 'hp', 'pro',
                          'kalo', 'amp', 'biar', 'bikin', 'bilang', 'gak', 'ga', 'krn', 'nya', 'nih', 'sih', 'si', 'tau', 'tdk', 'tuh', 'utk', 'ya', 'dong', 'redmi', 'note',
                          'jd', 'jgn', 'sdh', 'aja', 'n', 't', 'nyg', 'hehe', 'pen', 'u', 'nan', 'loh', 'rt', '&amp', 'yah', 'video', 'kelompok','buat', 'pake', 'alhamdulillah',
                          'terima kasih', 'terimakasih', 'terima', 'kasih', 'saya', 'materi', 'sepei', 'pertanyaan', 'peanyaannya', 'min', 'mint', 'minn', 'jadi', 'sama', 'udah', 'mungkin', 'mana'])

    filtered = [txt for txt in text if txt not in listStopwords]
    return filtered

data_copy = data2.copy()
data_copy['Comment'] = data_copy['Comment'].apply(lambda x: filteringText(tokenizingText(clean_text(x))))
```

```python
processed_data = {
    'Sentence':data_copy.Comment,
    'Sentiment':data_copy['Sentiment']
}

processed_data = pd.DataFrame(processed_data)
processed_data.head()
```

```python
processed_data['Sentence'] = processed_data['Sentence'].apply(lambda x: ' '.join(x))
processed_data.head()
```

```python
from wordcloud import WordCloud, STOPWORDS
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
sentiment_counts = processed_data['Sentiment'].value_counts()
```

```python
sentiments = sentiment_counts.index
counts = sentiment_counts.values
color_dict = {'Positive': 'green', 'Neutral': 'blue', 'Negative': 'red'}

colors = [color_dict[sentiment] for sentiment in sentiments]
plt.figure(figsize=(8, 6))
plt.bar(sentiments, counts, color=colors)
plt.title('Jumlah Data Berdasarkan Sentimen')
plt.xlabel('Sentimen')
plt.ylabel('Jumlah Data')
```

```python
# Concatenate all sentences into a single text
text = ' '.join(processed_data['Sentence'])

wordcloud = WordCloud(width=1600, height=800, random_state=21, max_font_size=110, collocations=False)

wordcloud.generate(text)

plt.figure(figsize=(10, 10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
